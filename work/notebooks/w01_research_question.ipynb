## Research Question

### Decision
The objective of this project is to determine which web pages should be prioritized for content refresh based on their content and search performance indicators.

### Action
The recommendations generated by the model can help SEO specialists, content managers, and website administrators decide which pages should be updated, improved, merged, or reviewed first.

### Cost of a Wrong Recommendation
If the model fails to identify pages that are actually declining, those pages may continue losing search visibility, organic traffic, and user engagement. On the other hand, recommending healthy pages for refresh may waste valuable time and resources.

### Why Machine Learning?
Machine learning can analyze many content and performance features simultaneously and recognize patterns that are difficult to detect using manual rules. The goal is to support better content prioritization through data-driven decision-making rather than relying on intuition alone.

## Quick Look at the Data

The starter dataset contains 30,000 web pages with multiple search-performance features. It includes several trend categories such as up, down, stable, flat, and new, making it suitable for identifying pages that may require content refresh. The dataset also contains useful numerical variables such as CTR, word count, and search volume, providing a solid foundation for building a content opportunity scoring model.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total Pages:", len(df))

print("\nTrend Direction Counts:")
print(df["trend_direction"].value_counts())

print("\nAverage CTR:")
print(round(df["ctr"].mean(), 3))

print("\nMedian Word Count:")
print(df["word_count"].median())
```
